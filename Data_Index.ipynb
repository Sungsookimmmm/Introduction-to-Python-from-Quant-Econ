```python
# the df.set_index method allows us to use one or more of the DataFrame's columns as the index
# first, create the DataFrame
df_year = df.set_index(["year"])
df_year.head()
```

```python
df_year.loc[2010]
```

```python
# To compute the difference in the average of all our variables from one year to the next.
df_year.loc[2009].mean(numeric_only=True) - df_year.loc[2008].mean(numeric_only=True)
```

```python
#After computing .mean(), the row labels(index) were the former column names
# These clumn names were used to align data when we wanted to ask pandas to compute the difference.

```

```python
df_year.loc[df_year["country"]=="United States","GDP"].loc[2010]
```

```python
df_year.loc[df_year["country"].isin(["United Kingdom","Germany"]),"GDP"].loc[2010]
```

```python
nx = df_year["Exports"] - df_year["Imports"]
nx.head(19)
```

```python
#Suppose that we accidentally had a bug in our code that swapped the data for Canada and Germany's net exports in 2017.
```

```python
ca17 = nx.iloc[[0]]
g17 = nx.iloc[[18]]
nx.iloc[[0]] = g17
nx.iloc[[18]] = ca17
nx.head(19)
```

```python
#If we now add nx to the DataFrame and compute investment
df_year["NetExports"] = nx
df_year["Investment"] = df_year.eval("GDP - Consumption - GovExpend - NetExports")
df_year.head(19)
```
