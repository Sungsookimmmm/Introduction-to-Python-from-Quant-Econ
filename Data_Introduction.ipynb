```python
# Manually computing Pearson correlation between West and MidWest unemployment rates

import numpy as np

# Step 1: Define the data
west = np.array([6.6, 6.0, 5.2, 4.6, 6.5, 5.5, 4.5, 8.6, 10.7, 8.5, 6.1])
midwest = np.array([4.5, 4.3, 3.6, 4.0, 5.7, 5.7, 4.9, 8.1, 8.7, 7.4, 5.1])

# Step 2: Compute the means
mean_west = np.mean(west)
mean_midwest = np.mean(midwest)

# Step 3: Compute deviations from the mean
dev_west = west - mean_west
dev_midwest = midwest - mean_midwest

# Step 4: Multiply corresponding deviations (covariance terms)
cov_terms = dev_west * dev_midwest

# Step 5: Compute covariance (sample)
covariance = np.sum(cov_terms) / (len(west) - 1)

# Step 6: Compute standard deviations (sample)
std_west = np.std(west, ddof=1)
std_midwest = np.std(midwest, ddof=1)

# Step 7: Compute Pearson correlation coefficient
correlation = covariance / (std_west * std_midwest)

# Print all intermediate results
print(f"Mean (West): {mean_west:.4f}")
print(f"Mean (MidWest): {mean_midwest:.4f}")
print(f"Deviations (West): {dev_west}")
print(f"Deviations (MidWest): {dev_midwest}")
print(f"Covariance terms: {cov_terms}")
print(f"Covariance: {covariance:.4f}")
print(f"Standard Deviation (West): {std_west:.4f}")
print(f"Standard Deviation (MidWest): {std_midwest:.4f}")
print(f"Pearson Correlation Coefficient: {correlation:.4f}")
```
